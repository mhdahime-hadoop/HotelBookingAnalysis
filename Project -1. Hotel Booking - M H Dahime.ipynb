```python
cancel_reser_perc = df['is_canceled'].value_counts(normalize = True)
cancel_reser_perc
```

```python
plt.figure(figsize = (5,4))
plt.title("Reservation Status Count")
plt.bar(['Not Canceled','Canceled'], df['is_canceled'].value_counts(), edgecolor = 'b', width = 0.7)
plt.show()
```

```python
plt.figure(figsize = (8,4))
axl = sns.countplot(x = 'hotel', hue = 'is_canceled', data = df, palette = 'Blues')
legend_labels,_ = axl. get_legend_handles_labels()
axl.legend(bbox_to_anchor=(1,1))
plt.title('Hotel wise Reservation Status', size = 25)
plt.xlabel('Hotel Name')
plt.ylabel('Numbers of Reservations')
plt.legend(['not canceled','canceled'])
plt.show()
```

```python
df.columns
```

```python
resort = df[df['hotel'] == 'Resort Hotel']
resort['is_canceled'].value_counts(normalize = True)
```

```python
city = df[df['hotel'] == 'City Hotel']
city['is_canceled'].value_counts(normalize = True)
```

```python
resort = resort.groupby('reservation_status_date')[['adr']].mean()
city = city.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize = (25,8))
plt.title('Hotel wise Average Daily Rates', fontsize = 35)
plt.plot(resort.index, resort['adr'], label = "Resort Hotel")
plt.plot(city.index, city['adr'], label = "City Hotel")
plt.legend(fontsize = 20)
plt.show()
```

```python
df['month'] = df['reservation_status_date'].dt.month
```

```python
plt.figure(figsize = (16,8))
axl = sns.countplot(x = 'month', hue = 'is_canceled', data = df, palette = 'bright')
legend_labels,_ = axl. get_legend_handles_labels()
axl.legend(bbox_to_anchor = (1,1))
plt.title('Month wise Reservation Status')
plt.xlabel('Months')
plt.ylabel('Numbers of Reservations')
plt.legend(['not canceled','canceled'])
plt.show()
```

```python
canceled_data = df[df['is_canceled'] == 1]
top_5_countries = canceled_data['country'].value_counts()[:5]
plt.figure(figsize = (8,8))
plt.title('Top 5 countries with canceled data')
plt.pie(top_5_countries, autopct = '%.2f', labels = top_5_countries.index)
plt.show()
```

```python
canceled_df_adr = canceled_data.groupby('reservation_status_date')[['adr']].mean()
canceled_df_adr.reset_index(inplace = True)
canceled_df_adr.sort_values('reservation_status_date', inplace = True)

not_canceled_data = df[df['is_canceled'] == 0]
not_canceled_data = not_canceled_data.groupby('reservation_status_date')[['adr']].mean()
not_canceled_data.reset_index(inplace = True)
not_canceled_data.sort_values('reservation_status_date', inplace = True)
```

```python
# visualization
```

```python
plt.figure(figsize = (20,6))
plt.title('Average Daily Rate')
plt.plot(not_canceled_data['reservation_status_date'], not_canceled_data['adr'], label = 'Not Canceled')
plt.plot(canceled_data['reservation_status_date'], canceled_data['adr'], label = 'Canceled')
plt.legend()
```

```python
plt.figure(figsize = (20,6))
plt.title('Average Daily Rate')
plt.plot(not_canceled_data['reservation_status_date'], not_canceled_data['adr'], label = 'Not Canceled')
plt.plot(canceled_data['reservation_status_date'], canceled_data['adr'], label = 'Canceled', alpha = .5)
plt.legend()
```
